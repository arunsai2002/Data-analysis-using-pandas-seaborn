```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

```python
df = pd.read_csv('ICICIBANK.csv',encoding= 'unicode_escape')
#unicode escape is used to avoid errors
#read_csv function loads & read the csv file
```

```python
df.shape
#shape keyword is used to check whether csv file is loaded or not
#it shows number of rows and columns in file
```

```python
df.head()
#if head() is used without no of rows, by default it shows first 5 rows
```

```python
df.head(10)
#head(n) function returns first n rows
```

```python
df.tail(10)
#tail(n) returns last n rows
```

```python
df.info()
#info method gives details about no of rows & columns, data type of every column & null or non-values in columns.
```

```python
df.drop(['VWAP','Symbol'],axis=1,inplace=True)
#to delete rows or columns in file, drop() method is used.
#axis = 1 , means it defines columns to be deleted
#if rows have to be deleted axis keyword is not used in the drop function.
#inplace = True means the data manipulated have to be saved.
```

```python
pd.isnull(df)
#isnull returns true if there are null values in file, else false.
```

```python
pd.isnull(df).sum()
#return number of null values column wise in file
```

```python
df.dropna(inplace=True)
#dropna method removes null values from file
```

```python
df.sort_values("Date",axis=0,ascending=True,inplace=True)
#sort_values method sorts data based on specified column, ascending or descending.
```

```python
df.info()
```

```python
df.shape
```

```python
df['Trades'] = df['Trades'].astype('int')
# astype used to change datatype of column selected.
```

```python
df.info()
```

```python
df['Trades'].dtypes
#dtypes define the datatype of selected column
```

```python
df.rename(columns={'%Deliverble':'percentage_delivery'})
#rename method renames column name/s.
```

```python
df.astype({'Last':'int', 'Close':'int'})
#to change data type of multiple columns
```

```python
df.info()
```

```python
df.describe()
#describe method shows all statistical calculations like mean, standard deviation, min, max, count, etc.. of every column
```

```python
#DATA VISUALIZATION USING PYTHON
```

```python
#countplot using seaborn
sns.countplot(x='Volume',data = df)
#here x axis is defined, defining y axis is not required. y axis gives count of x axis value automatically.

```

```python
df.groupby(['Date'],as_index = False)['Volume'].sum().sort_values(by='Volume',ascending=False)

#groupby groups columns based on some criteria.
#for e.g: in the above table, Volume is returned for particular DATE.
```

```python
groupbyvalue = df.groupby(['Date'],as_index = False)['Volume'].sum().sort_values(by='Volume',ascending=False)
sns.barplot(x='Date',y='Volume',data = groupbyvalue)

#barplot of Volume vs Date
```

```python
groupbyvalue = df.groupby(['Date'],as_index = False)['Trades'].sum().sort_values(by='Trades',ascending=False)
sns.barplot(x='Date',y='Trades',data = groupbyvalue)

#barplot of Trades vs Date
```

```python
groupbyvalue = df.groupby(['Date'],as_index = False)['Turnover'].sum().sort_values(by='Turnover',ascending=False)
sns.barplot(x='Date',y='Turnover',data = groupbyvalue)

#barplot of Turnover vs Date
```

```python
sns.barplot(x='Turnover',y='Volume',data = df)
sns.set(rc={'figure.figsize':(15,5)})
#set function sets width & height of plot

#barplot of Turnover vs Volume
```
